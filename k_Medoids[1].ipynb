```python
import pandas as pd
import numpy as np
import random
df= pd.read_csv("Hospitals.csv")#dataset
def dismilarity(x,y):
     return np.sqrt(np.sum(np.square(x-y)))#eucidean distance
lis1=[]
dist1=[]
dist2=[]
lis2=[]
indi=0
lis3=[]
dist3=[]
lis4=[]
dist4=[]
lis5=[]
dist5=[]
s=[]
ndf=df.sample(frac=4/len(df)) #taking for k=4 first
for i in range(100):
#print(ndf)
    for i in range(len(df)):
        #forming clusters first
        min_v=[]
        for j in range(len(ndf)):
            v=dismilarity(df.iloc[i],ndf.iloc[j] )
            min_v.append(v)
        if(min_v.index(min(min_v))==0):
            lis1.append(i)
                #print(lis1)
        elif(min_v.index(min(min_v))==1):
            lis2.append(i)
                #print(lis2)
        elif(min_v.index(min(min_v))==2):
            lis3.append(i)
        elif(min_v.index(min(min_v))==3):
            lis4.append(i)
    #now taking ad calculating cost values
    for i in ((lis1)):
        dis=0
        for j in ((lis1)):
            dis+=dismilarity(df.iloc[i],df.iloc[j] )
        dist1.append(dis)
    for i in ((lis2)):
        dis=0
        for j in ((lis2)):
            dis+=dismilarity(df.iloc[i],df.iloc[j] )
        dist2.append(dis)
    for i in ((lis3)):
        dis=0
        for j in ((lis3)):
            dis+=dismilarity(df.iloc[i],df.iloc[j] )
        dist3.append(dis)
    for i in ((lis4)):
        dis=0
        for j in ((lis4)):
            dis+=dismilarity(df.iloc[i],df.iloc[j] )
        dist4.append(dis)
    #print(dist1,"\n",dist2,"\n",dist3) 
    #calculating cluster heads
    indi1=dist1.index(min(dist1))
    cl_h1=lis1[indi1]
    indi2=dist2.index(min(dist2))
    cl_h2=lis2[indi2]
    indi3=dist3.index(min(dist3))
    cl_h3=lis3[indi3]
    indi4=dist4.index(min(dist4))
    cl_h4=lis4[indi4]
    new_df=[]
    new_df.append(cl_h1)
    new_df.append(cl_h2)
    new_df.append(cl_h3)
    new_df.append(cl_h4)
    lis5=lis1
    lis6=lis2
    lis7=lis3
    lis8=lis4
    lis1.clear()
    lis2.clear()
    lis3.clear()
    lis4.clear()
    for i in range(len(df)):
        nmin_v=[]
        for j in range(len(new_df)):
            v=dismilarity(df.iloc[i],df.iloc[new_df[j]] )
            nmin_v.append(v)
        if(nmin_v.index(min(nmin_v))==0):
            df.at[i,'Cluster_ID']=1
            lis1.append(i)
        elif(nmin_v.index(min(nmin_v))==1):
            df.at[i,'Cluster_ID']=2
            lis2.append(i)
        elif(nmin_v.index(min(nmin_v))==2):
            df.at[i,'Cluster_ID']=3
            lis3.append(i)
        elif(nmin_v.index(min(nmin_v))==3):
            df.at[i,'Cluster_ID']=4
            lis4.append(i)
    if(lis1==lis5 and lis2==lis6 and lis3==lis7 and lis4==lis8):
        break
#print(lis1,"\n",lis2,"\n",lis3,"\n",lis4)#final cluster formations
y=df.to_csv('clusters_k4.csv')
```
